```python
import pandas as pd

# Load Excel file
df = pd.read_excel("data/orders.xlsx")
```

```python
# Display information about the dataset
print(df.info())
```

```python
# Display the column names
print("\nColumn Names:")
print(df.columns)
```

```python
# Display the number of rows and columns
print("\nShape of Dataset:")
print(df.shape)
```

```python
# Count missing values in each column
print(df.isnull().sum())
```

```python
# Fill missing CouponCode values
df["CouponCode"] = df["CouponCode"].fillna("No Coupon")
```

```python
# Fill missing ReferralSource values
df["ReferralSource"] = df["ReferralSource"].fillna("Unknown")
```

```python
# Count duplicate rows
print("Duplicate Rows Before Cleaning:", df.duplicated().sum())
```

```python
# Count duplicate Order IDs
print("Duplicate Order IDs Before:", df["OrderID"].duplicated().sum())
```

```python
# Convert the Date column into datetime format
# Invalid dates will become NaT (Not a Time)
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
```

```python
# List of text columns
text_columns = [
    "Product",
    "ShippingAddress",
    "PaymentMethod",
    "OrderStatus",
    "TrackingNumber",
    "CouponCode",
    "ReferralSource"
]

# Remove extra spaces from text columns
for col in text_columns:
    df[col] = df[col].astype(str).str.strip()
```

```python
# List of numeric columns
numeric_columns = [
    "Quantity",
    "UnitPrice",
    "ItemsInCart",
    "TotalPrice"
]

# Convert columns to numeric datatype
for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python

# Check missing values after cleaning
print(df.isnull().sum())
```

```python
# Display final dataset information
print(df.info())
```

```python
# Display first 5 cleaned rows
print(df.head())
```

```python
# Save the cleaned dataset as a new Excel file
df.to_excel("cleaned_orders.xlsx", index=False)

print("Data Cleaning Completed Successfully!")
```
